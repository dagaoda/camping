```python
import numpy as np
import pandas as pd
import folium
```

```python
캠핑 = pd.read_csv('안산시_캠핑장.csv', encoding='cp949')
캠핑.head()
캠핑. tail()
```

```python
캠핑.isnull()
캠핑.isnull().sum()
```

```python
캠핑[['캠핑장명']].value_counts()
```

```python
캠핑.iloc[0]['캠핑장명']
캠핑.iloc[0]['위도']
캠핑.iloc[0]['경도']
```

```python
캠핑.shape
```

```python
for i in range(캠핑.shape[0]):
    print(캠핑.iloc[i]['캠핑장명'], 캠핑.iloc[i]['위도'], 캠핑.iloc[i]['경도'])
```

```python
m = folium.Map(
    location = [37.30604006894487, 126.85135837219448],
    zoom_start = 13
)

tooltip = 'Click!!'

for i in range(캠핑.shape[0]):
    folium.Marker(
        [캠핑.iloc[i]['위도'], 캠핑.iloc[i]['경도']],
        popup = f'<strong>{캠핑.iloc[i]["캠핑장명"]}</strong>',
        tooltip = tooltip
    ).add_to(m)

m
```

```python
from folium.features import CustomIcon

m = folium.Map(
    location=[37.1863447, 126.6311414],
    zoom_start=11
)

tooltip = 'Click!!'

for i in range(캠핑.shape[0]):
    icon_image = 'camping.png'
    icon = CustomIcon(
        icon_image,
        icon_size = (25, 25),
        popup_anchor = (30, -30),
    )
    
    folium.Marker(
        [캠핑.iloc[i]['위도'], 캠핑.iloc[i]['경도']],
        popup = f'<div style="width:300px"><strong>{캠핑.iloc[i]["캠핑장명"]}</strong><br>\
        전화번호 : {캠핑.iloc[i]["전화번호"]}<br>\
        홈페이지 : {캠핑.iloc[i]["홈페이지"]}<br>\
        주소 : {캠핑.iloc[i]["주소"]}<br>\
        <a href="http://grancity-xi.kr/">자료제공:그랑자이세종 장석호</a></div>',
        tooltip = tooltip,
        icon = icon
    ).add_to(m)

m
```

```python
m.save('camping.php')
```
